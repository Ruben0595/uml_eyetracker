```python
import matplotlib.pyplot as plt
```

```python
method = 'majority_vote'
task = 'Ball_Catch'
participant = 1

vote_treshold = 4
give_zero_lower_prio = 3

with open('./generated_data/majority_vote/%s/min label/participant_%s_vt_%s_zlp_%s.txt'%(task, participant, vote_treshold ,give_zero_lower_prio)) as f:
    labels = f.read()

labels = labels[1:-1].split(',')
for enum, label in enumerate(labels):
    labels[enum] = int(label)

totalframes = {}
totalframes['f'] = labels.count(1)
totalframes['gp'] = labels.count(2)
totalframes['s'] = labels.count(3)
totalframes['gf'] = labels.count(5)
```

```python
fig, ax = plt.subplots()
prevcount = (0,0,0,0)



D = totalframes
print(totalframes)
names, counts = zip(*D.items())
ax.bar(names, counts, label = ('reduce zero: ' + str(give_zero_lower_prio)), bottom=prevcount)


#vote_treshold = 1
give_zero_lower_prio = 0

with open('./generated_data/majority_vote/%s/min label/participant_%s_vt_%s_zlp_%s.txt'%(task, participant, vote_treshold ,give_zero_lower_prio)) as f:
    labels = f.read()

labels = labels[1:-1].split(',')
for enum, label in enumerate(labels):
    labels[enum] = int(label)

totalframes = {}
totalframes['f'] = labels.count(1)
totalframes['gp'] = labels.count(2)
totalframes['s'] = labels.count(3)
totalframes['gf'] = labels.count(5)

D = totalframes
names, counts = zip(*D.items())
ax.bar(names, counts, label = 'reduce zero: 0', bottom=prevcount)

ax.legend()
ax.set_title('ball catch, participant %s, vote treshold = %s'%(participant, vote_treshold))


#prevcount = tuple(map(sum, zip(prevcount, counts)))
```
